```python
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')

dataset['train']
```

```python
# turn the dataset into a dataframe using to_pandas()

#loading data
df = dataset['train'].to_pandas()
```

```python
# getting unique values for a field

df.job_title_short.unique()
```

```python
#filtering for a certain condition

df[df.job_title_short == 'Data Analyst']
```

```python
# multiple conditions using & operator

df[(df.job_title_short == 'Data Analyst') & (df.salary_year_avg > 100000)]
```

```python
# reassigning column to datetime data type

df['job_posted_date'] = pd.to_datetime(df.job_posted_date)

# job_posted_date is now datetime

df.info()
```

```python
# adding a column for the MONTH number for each 'job_posted_date'

df['job_posted_month'] = df.job_posted_date.dt.month

df.info()
```

```python
# sorting values in dataframe

df.sort_values(by = 'job_posted_date', inplace = True)

df
```
