```python
# using copy() with dataframes

dataset = load_dataset('lukebarousse/data_jobs')
df_orginal = dataset['train'].to_pandas()
```

```python
# using copy() ensures that the two dataframes (df_original and df_alterted) do not have the same unique id
# otherwise, changes to df_altered would also be applied to df_original

df_altered = df_orginal.copy()
```

```python
print('ID of df_original:              ', id(df_orginal))
print('ID of df_altered:               ', id(df_altered))
print('Are the two dataframes the same?', id(df_orginal) == id(df_altered))
```
