## What are the most demanded skills for the top 3 most popular data roles?

### Methodology

1. Clean-up skill column
2. Calculate skill count based on `job_title_short`
3. Calculate skill percentage
4. Plot final findings

```python
# import libraries

import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading data

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# data cleanup

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

# use apply() to clean 'job_skills' >> makes the column a list datatype instead of string

df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else(x))
```

```python
# filter dataframe specific to United States

df_US = df[df['job_country'] == 'United States']
```

```python
# breakout list of job_skills

df_skills = df_US.explode('job_skills')

df_skills[['job_title', 'job_skills']]
```

```python
# grouping

df_skills_count = df_skills.groupby(['job_skills', 'job_title_short']).size()

# converting from series to dataframe

df_skills_count = df_skills_count.reset_index(name='skill_count')

df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True)

df_skills_count
```

```python
# finding top 3 roles (convert from array to list)

job_titles = df_skills_count['job_title_short'].unique().tolist()

job_titles = sorted(job_titles[:3])

job_titles
```

```python
# begin plotting

fig, ax = plt.subplots(len(job_titles), 1)

# iterating through

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_count[df_skills_count['job_title_short'] == job_title].head(5)
    df_plot.plot(kind='barh', x='job_skills', y='skill_count', ax=ax[i], title=job_title)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].legend().set_visible(False)

fig.suptitle('Count of Top 5 Skills in Job Postings', fontsize=15)
plt.tight_layout(h_pad=0.5) # fixes overlap
plt.show()
```

```python
# total counts of jobs based on title, convert from series to dataframe

df_job_title_count = df_US['job_title_short'].value_counts().reset_index(name='jobs_total')
```

```python
# merging dataframes (left join)

df_skills_perc = pd.merge(df_skills_count, df_job_title_count, how='left', on='job_title_short')

df_skills_perc['skill_percent'] = 100 * df_skills_perc['skill_count'] / df_skills_perc['jobs_total']

df_skills_perc
```

```python
fig, ax = plt.subplots(len(job_titles), 1)

sns.set_theme(style='ticks')

# iterating through

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_perc[df_skills_perc['job_title_short'] == job_title].head(5)
    sns.barplot(data=df_plot, x='skill_percent', y='job_skills', ax=ax[i], hue='skill_count', palette='dark:b_r')
    ax[i].set_title(job_title)
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].legend().remove()
    ax[i].set_xlim(0, 78)

    # labeling graph

    for n, v in enumerate(df_plot['skill_percent']):
        ax[i].text(v + 1, n, f'{v:.0f}%', va='center')

    # removing repetitive x axes
    if i != len(job_titles) - 1:
        ax[i].set_xticks([])

fig.suptitle('Likelihood of Skills Requested in US Job Postings', fontsize=15)
plt.tight_layout(h_pad=0.5) # fixes overlap
plt.show()
```
